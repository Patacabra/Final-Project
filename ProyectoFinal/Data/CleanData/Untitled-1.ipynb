```python
import pandas as pd
import numpy as np
```

```python
c = pd.read_csv('coursera.csv')
u = pd.read_csv('udemy.csv')
d = pd.read_csv('domestika.csv')
```

```python
cc = pd.Series("Coursera", index=c.index)

c = c.assign(Web=cc)
```

```python
c = c.drop(columns=['Unnamed: 0'])
```

```python
c
```

```python
cu = pd.Series("Udemy", index=u.index)

u = u.assign(Web=cu)
```

```python
u
```

```python
cd = pd.Series("Domestika", index=d.index)

d = d.assign(Web=cd)
```

```python
d = d.rename(columns={'course_title': 'Title', 'course_organization':'Organization','course_price': 'Course_price','course_students_enrolled': 'Course_students_enrolled', 'course_description':'Course_description'})
```

```python
d = d.drop(columns=['Unnamed: 0', 'course_Certificate_type', 'course_difficulty', 'course_rating'])
```

```python
d
```

```python
cursos = pd.concat([c, u, d], ignore_index=True)

print(cursos)
```

```python
cursos
```

```python
cursos.to_csv('cursos.csv', index=False)
```
