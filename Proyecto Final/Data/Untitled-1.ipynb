```python

import glob

# Obtener una lista de todos los archivos CSV que quieres concatenar
csv_files = glob.glob('udemyoffprodclean.csv') + glob.glob('udemydesignclean.csv') + glob.glob('udemydevelopclean.csv') + glob.glob('udemyitsoftclean.csv')

# Crear una lista de DataFrames
dfs = [pd.read_csv(file) for file in csv_files]

# Concatenar los DataFrames en un solo DataFrame
concatenated_df = pd.concat(dfs, ignore_index=True)

# Escribir el DataFrame concatenado en un archivo CSV
concatenated_df.to_csv('resultados.csv', index=False)
```
