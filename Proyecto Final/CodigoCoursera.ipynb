```python
def scrape_coursera_data(driver, course_title, course_organization, course_Certificate_type, course_rating):
    count = 0
    while count < 84:
        titles = [e.text for e in driver.find_elements(By.TAG_NAME, 'h2')]
        if titles:
            course_title.extend(titles)
            time.sleep(3)


        organizations = [e.text for e in driver.find_elements(By.CLASS_NAME, 'cds-119.css-2fzscr.cds-121')]
        if organizations:
            course_organization.extend(organizations)
            time.sleep(3)

        certificate_types = [e.text for e in driver.find_elements(By.CLASS_NAME, 'cds-119.css-14d8ngk.cds-121')]
        if certificate_types:
            course_Certificate_type.extend(certificate_types)
            time.sleep(3)

        ratings = [e.text for e in driver.find_elements(By.CLASS_NAME, 'cds-119.css-zl0kzj.cds-121')]
        if ratings:
            course_rating.extend(ratings)
            time.sleep(3)

        try:
            next_page = driver.find_element(By.XPATH, '//*[@id="main"]/div[2]/div[1]/div/div/div/div/div/div/div[2]/div/div/div/div[1]/div/button[7]').click()
            time.sleep(3)
            count += 1
        except:
            break
```
