```python
import time

from undetected_chromedriver import Chrome, ChromeOptions
from selenium.common.exceptions import WebDriverException
from selenium.webdriver.remote.webdriver import By
import selenium.webdriver.support.expected_conditions as EC  # noqa
from selenium.webdriver.support.wait import WebDriverWait

import undetected_chromedriver as uc
options = ChromeOptions()
#options.add_argument("--headless")
options.add_argument("--no-sandbox")
options.add_argument('--start-maximized')  
options.add_argument("--disable-dev-shm-usage")
options.add_argument("--disable-extensions")
options.add_argument("--disable-gpu")
options.add_argument("--disable-setuid-sandbox")
options.add_argument("--disable-remote-fonts")

with Chrome(options=options) as driver:
    driver.get('https://www.udemy.com/es/courses/development/?sort=newest')
    title = driver.title
    print(title)
```

```python
time.sleep(60)
cookies = driver.find_element(By.XPATH, '//*[@id="onetrust-accept-btn-handler"]').click()
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def scrape_udemy_data(driver, course_title1, course_organization1, course_price1, course_info1):
    page_number = 1
    while page_number <= 20:
        titles = [e.text for e in driver.find_elements(By.TAG_NAME, 'h3')]
        if titles:
            course_title1.extend(titles)
            time.sleep(3)

        organizations = [e.text for e in driver.find_elements(By.CLASS_NAME, 'course-card--instructor-list--nH1OC')]
        if organizations:
            course_organization1.extend(organizations)
            time.sleep(3)

        course_prices = [e.text for e in driver.find_elements(By.CLASS_NAME, 'price-text--container--103D9.course-card--price-text-container--XIYmk')]
        if course_prices:
            course_price1.extend(course_prices)
            time.sleep(3)

        info = [e.text for e in driver.find_elements(By.CLASS_NAME, 'course-card--row--29Y0w')]
        if info:
            course_info1.extend(info)
            time.sleep(3)

        # Build the URL for the next page
        page_number += 1
        next_page_url = "https://www.udemy.com/es/courses/development/?p={}&sort=newest".format(page_number)

        # Navigate to the next page
        driver.get(next_page_url)

        # Wait for the next page to load
        try:
            WebDriverWait(driver, 40).until(EC.presence_of_element_located((By.TAG_NAME, 'h3')))
        except:
            print("Timeout: Failed to load the page.")
            break

        time.sleep(3)
```

```python
resultado3 = []
termino_en_total = False
ultimo_indice_calificacion = -1  # inicializar con un índice no válido

for i in range(len(course_info1)):
    if 'Calificación' in course_info1[i]:
        resultado3.append(course_info1[i])
        ultimo_indice_calificacion = i
        termino_en_total = False
    elif 'en total' in course_info1[i]:
        if ultimo_indice_calificacion == -1 or ultimo_indice_calificacion <= i:
            resultado3.append(course_info1[i])
            termino_en_total = True
        else:
            # agregar la cadena actual antes de la última Calificación
            resultado3.insert(ultimo_indice_calificacion, course_info1[i])
            ultimo_indice_calificacion += 1
    elif ultimo_indice_calificacion > i:
        # agregar la cadena actual después de la última Calificación
        resultado3.insert(ultimo_indice_calificacion + 1, course_info1[i])
        ultimo_indice_calificacion += 1
    else:
        termino_en_total = False

resultado3
```
